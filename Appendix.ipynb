### The coding details of principal component (PC) analysis based on the genomic SNP data and visualization using a customized Python script

```python
# Loading package
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from matplotlib import cm
import seaborn as sns; sns.set()
import matplotlib as mlp
import sklearn
from sklearn.decomposition import PCA
```

```python
# Package version
print("pandas",pd.__version__,
"numpy",np.__version__,
"matplotlib",mpl.__version__,
"seaborn",sns.__version__,
"sklearn", sklearn.__version__)
```

```python
# Loading data for PopFM
pheno = pd.read_table("./GS2018/Lin2018_phenotype.txt") # Phenotypes
I = pd.read_table('./GS2018/reorderG.txt',sep="\s+")-1 # Genotypes
bcw=pheno["count"] # Heterobothrium okamotoi count
```

```python
# PC analysis
pca = PCA()
X_r = pca.fit(I).transform(I)
```

```python
# Visualization
lin_wide=1.28
font_size=12
s_value=25
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.subplot(2, 1, 1)
plt.style.use('seaborn-white')
plt.rc('font',size=font_size)
plt.rc('lines',lw=lin_wide)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.sans-serif'] = 'Times New Roman'
plt.rcParams['axes.linewidth'] = lin_wide
plt.rcParams['xtick.major.width'] = lin_wide
plt.rcParams['xtick.minor.width'] = lin_wide
plt.rcParams['xtick.labelsize'] = font_size
plt.rcParams['ytick.major.width'] = lin_wide
plt.rcParams['ytick.minor.width'] = lin_wide
plt.rcParams['ytick.labelsize'] = font_size
plt.rcParams['axes.spines.bottom'] = True
# PC1 & PC2
plt.scatter(X_r[:, 0], X_r[:, 1], edgecolor='k',cmap=cm.Reds,linewidth=lin_wide,s=s_value)
plt.xlabel('PC%s (%.1f%%)' % (1, np.round(pca.explained_variance_ratio_[0],3)*100),size=font_size)
plt.ylabel('PC%s (%.1f%%)' % (2, np.round(pca.explained_variance_ratio_[1],3)*100),size=font_size)
plt.subplot(2, 1, 2)
# PC3 & PC4
plt.scatter(X_r[:, 2], X_r[:, 3], edgecolor='k',cmap=cm.Reds,linewidth=lin_wide,s=s_value)
plt.xlabel('PC%s (%.1f%%)' % (3, np.round(pca.explained_variance_ratio_[2],3)*100),size=font_size)
plt.ylabel('PC%s (%.1f%%)' % (4, np.round(pca.explained_variance_ratio_[3],3)*100),size=font_size)
plt.gcf().set_size_inches(7, 12)
# plt.savefig("PCA1.pdf",format="pdf")
plt.show()
```
